```python
df = pd.read_csv("Dataset/Steel_industry_data.csv")
```

```python
df = df[['date','Usage_kWh']]
```

```python
df["date"] = pd.to_datetime(df["date"], format = "%d/%m/%Y %H:%M", errors = "coerce")
```

```python
dfx = df.copy()
```

```python
df.index = df["date"]
df.drop(["date"],axis=1, inplace=True)
```

```python
df.info()
```

```python
df.shape
```

```python
scaler = MinMaxScaler()
scaler_data = scaler.fit_transform(df)
```

```python
index = np.arange(0, len(scaler_data), 1)
scaler_df = pd.DataFrame(scaler_data, index=index, columns=['Usage_kWh'])
```

```python
train_size = int(len(scaler_df) * 0.9)
train, test = scaler_df.iloc[0:train_size], scaler_df.iloc[train_size:len(scaler_df)]
```

```python
print("Train size:", len(train), "Test size:", len(test))
```

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)        
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
n_steps = 24
```

```python
X_train, y_train = create_dataset(train, train['Usage_kWh'], n_steps)
```

```python
X_test, y_test = create_dataset(test, test['Usage_kWh'], n_steps)
```

```python
print("Training data shape:", X_train.shape, y_train.shape)
print("Test data shape:", X_test.shape, y_test.shape)
```

```python
def statistical_ensemble_learning():
    model = Sequential()
    model.add(LSTM(units=128, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(units=64))
    model.add(Dropout(0.2))
    model.add( Dense(units=1))
    return model
```

```python
nn_model = statistical_ensemble_learning()
```

```python
nn_model.compile(
    loss='mean_squared_error',
    optimizer=Adam(0.001)
)
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
can_train = True
```

```python
if can_train:
    history = nn_model.fit(
        X_train, y_train,
        epochs=500,
        batch_size=24,
        validation_split=0.1,
        verbose=1,
        shuffle=False,
        callbacks=[early_stopping]
    )
    nn_model.save("Steel-Industry-Energy-Prediction-DS.keras")
else:
    model = load_model("Steel-Industry-Energy-Prediction-DS.keras")
```

```python
nn_pred = nn_model.predict(X_test)
df_nn_final = test[nn_pred.shape[0]*-1:].copy()
df_nn_final['Prediction'] = nn_pred[:,0]
df_reset_index = df.reset_index()
df_nn_final['Date'] = df_reset_index['date']
df_nn_final = df_nn_final.set_index(["Date"], drop=True)
```

```python
df_nn_final['Usage_kWh'] = scaler.inverse_transform(df_nn_final['Usage_kWh'].to_numpy().reshape(-1, 1))
df_nn_final['Prediction'] = scaler.inverse_transform(df_nn_final['Prediction'].to_numpy().reshape(-1, 1))
```

```python
train2 = train.copy()
train2.index = df.index[:len(train)]
train2['Usage_kWh'] = scaler.inverse_transform(train2['Usage_kWh'].to_numpy().reshape(-1, 1))
train2['Datetime'] = dfx['date']
train2 = train2.set_index(['Datetime'])
```

```python
def report_metrics(y_true, y_pred):
    print("Explained Variance:\n\t", explained_variance_score(y_true, y_pred))
    print("MAE:\n\t", mean_absolute_error(y_true, y_pred))
    print("RMSE:\n\t", np.sqrt(mean_squared_error(y_true, y_pred)))
```

```python
plt.figure(figsize=(50,15))
plt.title('Statistical Ensemble Learning Model')
plt.plot(train2, color='red', label='Actual Energy Usage (Train)')
plt.xlabel('DateTime')
plt.ylabel('Usage_kWh')
plt.plot(df_nn_final['Usage_kWh'], color='green', label='Actual Energy Usage (Test)')
plt.plot(df_nn_final['Prediction'], color='purple', label='Predicted Energy Usage (LSTM)')
plt.legend()

plt.show()

report_metrics(df_nn_final['Usage_kWh'].squeeze(), df_nn_final['Prediction'].squeeze())
```
